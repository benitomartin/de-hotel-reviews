```python
# Convert to datetime

df["Review_Date"] = pd.to_datetime(df["Review_Date"])
df.info()    
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# Check missing values

def missing_values_summary(df):
    missing_counts = df.isnull().sum()
    total_count = df.shape[0]  # Total number of rows in the DataFrame

    missing_df = pd.DataFrame({'Column': missing_counts.index, 'Missing Count': missing_counts.values})
    missing_df['Missing Percentage'] = (missing_df['Missing Count'] / total_count) * 100

    return missing_df

missing_values_summary(df)
```

```python
# Remove missing values

df = df.dropna()
```

```python
# Changing United Kingdom to UK

df['Hotel_Address'] = df['Hotel_Address'].str.replace('United Kingdom', 'UK')
```

```python
list(set(df['Hotel_Address']))[:11]
```

```python
# Creating a new column with the country name, extracting the last word of the hotel adress column

df['Hotel_Country'] = df['Hotel_Address'].str.split().str[-1]
```
